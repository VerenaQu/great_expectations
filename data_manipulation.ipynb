# Data Manipulation

#### Import Pandas

```python
import pandas as pd
```

#### Read airline data

```python
df = pd.read_csv('../data/airline_data.csv')
```

#### Filter Southwest Airlines

```python
df_southwest = df.loc[df["UNIQUE_CARRIER_NAME"] == "Southwest Airlines Co."]
df_southwest.head(10)
```

##### Add seats per Flight

```python
df_southwest["SEATS_PER_FLIGHT"] = df_southwest["SEATS"] / df_southwest["DEPARTURES_PERFORMED"]
df_southwest.tail(15)
```

##### Add average passengers per flight

```python
df_southwest["AV_PASSENGERS_PER_FLIGHT"] = df_southwest["PASSENGERS"] / df_southwest["DEPARTURES_PERFORMED"]
df_southwest.tail(15)
```

##### Round average passengers per flight

```python
df_southwest['AV_PASSENGERS_PER_FLIGHT'] = df_southwest['AV_PASSENGERS_PER_FLIGHT'].apply(lambda x: round(x, 2))
df_southwest.head(15)
```

#### Manipulate SEATS_PER_FLIGHT -> insert false numbers

```python
df_southwest.loc[1033:1035,'SEATS_PER_FLIGHT'] = 299
df_southwest.head()
```

#### Download dataset

```python
df_southwest.to_csv('southwest_passengers.csv', index=False)
```
